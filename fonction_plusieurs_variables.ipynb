# fonction de plusieurs variables

### Définition des fonctions de plusieurs variables

Une **fonction de plusieurs variables** est une fonction qui dépend de plusieurs entrées (ou arguments) et produit une sortie. Plus formellement, une fonction $ f $ de plusieurs variables est une relation mathématique qui associe un ou plusieurs points dans un espace multidimensionnel à une valeur scalaire ou vectorielle.

#### Notation générale
Soit $ f : D \subset \mathbb{R}^n \to \mathbb{R}^m $, où :
- $ D $ est un sous-ensemble du domaine $ \mathbb{R}^n $, appelé l'ensemble de définition.
- $ n $ est le nombre de variables d'entrée.
- $ m $ est le nombre de valeurs de sortie.

Si $ m = 1 $, la fonction est scalaire (elle retourne un seul nombre réel). Si $ m > 1 $, la fonction est vectorielle (elle retourne un vecteur).

### Exemples concrets

1. **Fonction scalaire de deux variables** :
   Une fonction $ f(x, y) $ prend deux variables en entrée ($ x $ et $ y $) et retourne un nombre réel :
   $$
   f(x, y) = x^2 + y^2
   $$

2. **Fonction vectorielle de trois variables** :
   Une fonction $ \mathbf{F}(x, y, z) $ peut prendre trois variables en entrée et retourner un vecteur :
   $$
   \mathbf{F}(x, y, z) = \begin{bmatrix} x + y \\ y + z \\ z + x \end{bmatrix}
   $$

3. **Exemple général** :
   Soit $ f(x_1, x_2, \dots, x_n) $ une fonction qui dépend de $ n $ variables. Par exemple :
   $$
   f(x_1, x_2, x_3) = x_1^2 + 2x_2^2 + 3x_3^2 - x_1x_2x_3
   $$

# Expression des distances de M à A, B et C en fonction de $(x, y)$

Nous allons dériver les fonctions de distance entre le point $M$ (avec les coordonnées $(x, y)$) et chaque sommet du triangle $ABC$. La distance entre deux points $(x_1, y_1)$ et $(x_2, y_2)$ est donnée par la formule de la distance euclidienne :

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

Définissons $M$ comme $(x, y)$, et utilisons les coordonnées fixes de $A$, $B$ et $C$ :

- **Distance de $M$ à $A$** ($A = (1, 1)$) :
  $$
  d_{MA}(x, y) = \sqrt{(x - 1)^2 + (y - 1)^2}
  $$

- **Distance de $M$ à $B$** ($B = (3, 3)$) :
  $$
  d_{MB}(x, y) = \sqrt{(x - 3)^2 + (y - 3)^2}
  $$

- **Distance de $M$ à $C$** ($C = (1, -1)$) :
  $$
  d_{MC}(x, y) = \sqrt{(x - 1)^2 + (y + 1)^2}
  $$

Ces fonctions expriment les distances entre $M$ et chaque sommet du triangle en fonction des variables $x$ et $y$, permettant ainsi à $M$ de se déplacer n'importe où dans le plan.

### Somme des distances au carré (optionnelle)

$$
S^2(x, y) = d_{MA}^2(x, y) + d_{MB}^2(x, y) + d_{MC}^2(x, y)
$$

En substituant les expressions des distances au carré :

$$
S^2(x, y) = \big((x - 1)^2 + (y - 1)^2\big) + \big((x - 3)^2 + (y - 3)^2\big) + \big((x - 1)^2 + (y + 1)^2\big)
$$

Développons cette expression :

1. Pour $ d_{MA}^2(x, y) $ :
   $$
   (x - 1)^2 + (y - 1)^2 = x^2 - 2x + 1 + y^2 - 2y + 1 = x^2 + y^2 - 2x - 2y + 2
   $$

2. Pour $ d_{MB}^2(x, y) $ :
   $$
   (x - 3)^2 + (y - 3)^2 = x^2 - 6x + 9 + y^2 - 6y + 9 = x^2 + y^2 - 6x - 6y + 18
   $$

3. Pour $ d_{MC}^2(x, y) $ :
   $$
   (x - 1)^2 + (y + 1)^2 = x^2 - 2x + 1 + y^2 + 2y + 1 = x^2 + y^2 - 2x + 2y + 2
   $$

Ajoutons ces trois termes ensemble :

$$
S^2(x, y) = \big(x^2 + y^2 - 2x - 2y + 2\big) + \big(x^2 + y^2 - 6x - 6y + 18\big) + \big(x^2 + y^2 - 2x + 2y + 2\big)
$$

Regroupons les termes similaires :

- Les termes en $ x^2 $ : $ x^2 + x^2 + x^2 = 3x^2 $
- Les termes en $ y^2 $ : $ y^2 + y^2 + y^2 = 3y^2 $
- Les termes en $ x $ : $ -2x - 6x - 2x = -10x $
- Les termes en $ y $ : $ -2y - 6y + 2y = -6y $
- Les constantes : $ 2 + 18 + 2 = 22 $

Ainsi, la somme des distances au carré est :

$$
S^2(x, y) = 3x^2 + 3y^2 - 10x - 6y + 22
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Définir la fonction S(x, y) (somme des distances)
def S(x, y):
    d_MA = (x - 1)**2 + (y - 1)**2
    d_MB = (x - 3)**2 + (y - 3)**2
    d_MC = (x - 1)**2 + (y + 1)**2
    return d_MA + d_MB + d_MC

# Définir la grille pour x et y
x = np.linspace(-2, 5, 500)  # Plage de valeurs pour x
y = np.linspace(-2, 5, 500)  # Plage de valeurs pour y
X, Y = np.meshgrid(x, y)     # Créer une grille de points (x, y)

# Calculer S(x, y) pour chaque point de la grille
Z = S(X, Y)

# Tracer la surface 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('Surface 3D de S(x, y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('S(x, y)')

# Tracer le contour plot
plt.subplot(1, 2, 2)
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot de S(x, y)')
plt.xlabel('x')
plt.ylabel('y')

# Ajouter les points A, B, C
points = [(1, 1), (3, 3), (1, -1)]
for point in points:
    plt.scatter(*point, color='red', s=100, label=f'Point {point}')

plt.legend()
plt.tight_layout()
plt.show()
```

# Optimisation 

![optimisation](./optimisation.png)

#### **Méthode itérative : Descente de gradient**
   - Si résoudre $ \nabla f(x) = 0 $ analytiquement est compliqué, on utilise la descente de gradient, qui suit la règle de mise à jour :
     $$
     x_{k+1} = x_k - \alpha \nabla f(x_k)
     $$
     où :
     - $ x_k $ est l'estimation courante du minimum.
     - $ \alpha > 0 $ est le taux d'apprentissage (ou pas de descente).
     - $ \nabla f(x_k) $ est le gradient de $ f(x) $ au point $ x_k $.

   - La descente de gradient converge vers le minimum global si $ f(x) $ est convexe et si $ \alpha $ est choisi correctement.


```python
import numpy as np
import matplotlib.pyplot as plt

# Calculer les dérivées partielles de S²(x, y)
def gradient_S_squared(x, y):
    # Dérivée par rapport à x
    dx = 2 * ((x - 1) + (x - 3) + (x - 1))
    # Dérivée par rapport à y
    dy = 2 * ((y - 1) + (y - 3) + (y + 1))
    return dx, dy

# Descente de gradient avec enregistrement des points intermédiaires
def gradient_descent(alpha=0.01, max_iterations=1000, tolerance=1e-6):
    # Initialiser x et y avec des valeurs aléatoires ou fixées
    x, y = 0, 0  # Point de départ
    trajectory = [(x, y)]  # Enregistrer la trajectoire
    
    for i in range(max_iterations):
        # Calculer le gradient
        dx, dy = gradient_S_squared(x, y)
        
        # Mettre à jour x et y
        x_new = x - alpha * dx
        y_new = y - alpha * dy
        
        # Vérifier la convergence
        if np.sqrt((x_new - x)**2 + (y_new - y)**2) < tolerance:
            print(f"Convergence atteinte après {i+1} itérations.")
            break
        
        # Mettre à jour les valeurs de x et y
        x, y = x_new, y_new
        trajectory.append((x, y))  # Enregistrer la nouvelle position
    
    return x, y, np.array(trajectory)

# Exécuter la descente de gradient
x_min, y_min, trajectory = gradient_descent(alpha=0.01, max_iterations=1000, tolerance=1e-6)

# Afficher le résultat
print(f"Minimum trouvé à : ({x_min:.4f}, {y_min:.4f})")
print(f"Valeur de S² au minimum : {S(x_min, y_min):.4f}")

# Tracer les courbes de niveau et la trajectoire
# Définir une grille pour x et y
x_vals = np.linspace(-1, 4, 500)
y_vals = np.linspace(-2, 4, 500)
X, Y = np.meshgrid(x_vals, y_vals)
Z = S_squared(X, Y)

# Créer le contour plot
plt.figure(figsize=(10, 8))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour, label='Valeur de S²(x, y)')
plt.contour(X, Y, Z, levels=50, colors='gray', alpha=0.5)

# Ajouter les points A, B, C
points = [(1, 1), (3, 3), (1, -1)]
for point in points:
    plt.scatter(*point, color='red', s=100, label=f'Point {point}')

# Tracer la trajectoire de la descente de gradient
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', color='blue', label='Trajectoire GD')
plt.scatter(trajectory[0, 0], trajectory[0, 1], color='green', s=100, label='Départ')
plt.scatter(x_min, y_min, color='black', s=100, label='Minimum')

# Ajouter des flèches pour le gradient à quelques points
for i in range(0, len(trajectory), 5):  # Prendre un échantillon tous les 5 points
    x_i, y_i = trajectory[i]
    dx, dy = gradient_S_squared(x_i, y_i)
    plt.quiver(x_i, y_i, dx, dy, angles='xy', scale_units='xy', scale=10, color='orange', width=0.005)

# Ajouter des labels et une légende
plt.title('Descente de gradient avec gradients perpendiculaires aux courbes de niveau')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

1. Si $ f(x) $ est une fonction convexe et différentiable, alors :
   $$
   \nabla f(x^*) = 0
   $$
   où $ x^* $ est le point qui minimise $ f(x) $.

2. En pratique, si la solution analytique (résolution directe de $ \nabla f(x) = 0 $) est impossible ou difficile à obtenir, on utilise des méthodes itératives comme la **descente de gradient** pour converger vers le minimum.

---

### Règle pour minimiser une fonction convexe

Pour minimiser une fonction convexe $ f(x) $, voici les étapes générales :

#### 1. **Condition nécessaire du premier ordre**
   - Une condition nécessaire pour que $ x^* $ soit un minimum local est que le gradient de $ f(x) $ soit nul :
     $$
     \nabla f(x^*) = 0
     $$

#### 2. **Condition suffisante pour une fonction convexe**
   - Si $ f(x) $ est convexe, alors cette condition ($ \nabla f(x^*) = 0 $) est aussi **suffisante** pour garantir que $ x^* $ est un minimum global.

### Application à notre cas : Minimisation de $ S^2(x, y) $

Dans notre problème, la fonction $ S^2(x, y) $ est une somme de carrés de distances, ce qui en fait une fonction **convexe** (car chaque terme $ (x - x_i)^2 + (y - y_i)^2 $ est convexe). Nous pouvons donc appliquer la règle ci-dessus.

#### 1. Condition nécessaire du premier ordre
La condition pour trouver le minimum est de résoudre :
$$
\nabla S^2(x, y) = 0
$$
où $ \nabla S^2(x, y) $ est le gradient de $ S^2(x, y) $.

#### 2. Calcul du gradient
La fonction $ S^2(x, y) $ est donnée par :
$$
S^2(x, y) = 3x^2 + 3y^2 - 10x - 6y + 22 
$$

Développons ses dérivées partielles :

- Dérivée par rapport à $ x $ :
  $$
  \frac{\partial S^2}{\partial x} =  6x - 10
  $$

- Dérivée par rapport à $ y $ :
  $$
  \frac{\partial S^2}{\partial y} =  6y - 6
  $$

Ainsi, le gradient est :
$$
\nabla S^2(x, y) = \begin{bmatrix} 6x - 10 \\ 6y - 6 \end{bmatrix}
$$

#### 3. Résolution analytique
Pour trouver le minimum, nous résolvons $ \nabla S^2(x, y) = 0 $ :
$$
6x - 10 = 0 \quad \Rightarrow \quad x = \frac{10}{6} = \frac{5}{3}
$$
$$
6y - 6 = 0 \quad \Rightarrow \quad y = \frac{6}{6} = 1
$$

Donc, le minimum est atteint en :
$$
(x^*, y^*) = \left(\frac{5}{3}, 1\right)
$$

---

### Vérification : Valeur de $ S^2(x^*, y^*) $
Substituons $ x = \frac{4}{3} $ et $ y = \frac{2}{3} $ dans $ S^2(x, y) $ :
$$
S^2\left(\frac{5}{3}, 1\right) = 3x^2 + 3y^2 - 10x - 6y + 22=3*(\frac{5}{3})^2+3-10*\frac{5}{3}-6+22=\frac{75+27-150-54+198}{9}
$$

Après calcul, on obtient :
$$
S^2\left(\frac{5}{3}, 1\right) = \frac{96}{9} \approx 10.66667
$$

---

### Conclusion

Le minimum de $ S^2(x, y) $ est atteint en :
$$
\boxed{(x^*, y^*) = \left(\frac{5}{3},1\right)}
$$

Cela correspond également au point de Fermat-Torricelli du triangle formé par $ A(1, 1) $, $ B(3, 3) $, et $ C(1, -1) $.
